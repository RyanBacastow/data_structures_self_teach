### Question 1
##### Read the input one line at a time and then write the lines out in reverse order, so that the last input line is printed first, then the second last input line, and so on

```python
%%time
#Q1: I answered this question using a deque implementation of the list interface, using pop() in python.
with open('new_lorem.txt', 'r+') as f:
    xlist = []
    for line in f.readlines():
        xlist.append(line)
    for y in range(0,len(xlist)):
        print(xlist.pop())
```

```python
%%time
#Q1 Alternative answer utilizing singular readline() instead of reading entire file into memory
with open('new_lorem.txt', 'r') as f:
    line = f.readline()
    my_list = []
    i = 1
    while line:
        my_list.append(str(i) + '. ' + line)
        line = f.readline()
        i+=1
    for y in range(0,len(my_list)):
        print(my_list.pop())
```

### Question 2
Read the first 50 lines of input and then write them out in reverse
order. Read the next 50 lines and then write them out in reverse
order. Do this until there are no more lines left to read, at which
point any remaining lines should be output in reverse order.
In other words, your output will start with the 50th line, then the
49th, then the 48th, and so on down to the first line. This will be
followed by the 100th line, followed by the 99th, and so on down to
the 51st line. And so on.
Your code should never have to store more than 50 lines at any given
time.

```python
%%time
def reset():
    i = 1
    my_list = []
    return i, my_list

with open('new_lorem.txt', 'r') as f:
    i, my_list= reset()
    line = f.readline()
    counter = 1
    master_list=[]
    while i < 51:
        if line:
            if i == 50:
                my_list.append(str(i) + '. ' + line)
                master_list.append(str(i) + '. ' + line)
                print("BLOCK {}".format(counter))
                counter += 1
                for y in range(0,len(my_list)):
                    print(my_list.pop())
                i, my_list = reset()
                line = f.readline()
            else:
                master_list.append(str(i) + '. ' + line)
                my_list.append(str(i) + '. ' + line)
                line = f.readline()
                i+=1
        else:
            print("BLOCK {}".format(counter))
            for y in range(0,len(my_list)):
                print(my_list.pop())
            print('DONE, TOTAL LINES: {}'.format(str(len(master_list))))
            break
```
